### Question 9
How many movies in the database were produced by Pixar Animation Studios?



```python
%%sql 
# Your code here
SELECT count(pc.production_company_name) AS "Movies Produced By Pixar"
FROM productioncompanies AS pc
INNER JOIN productioncompanymap AS pcm
ON pc.production_company_id = pcm.production_company_id
WHERE pc.production_company_name = "Pixar Animation Studios";
```

### Question 10
How many movies are in the database that are both a Romance and a Comedy?


```python
%%sql 
# Your code here
SELECT count(1) AS Movies
FROM (
    SELECT movie_id 
    FROM genremap
    WHERE genre_id IN (35, 10749)
    GROUP BY movie_id
    HAVING count(genre_id) = 2
) AS t
```

### Question 11
What is the most popular action movie that has some German in it? (Hint: The German word for German is Deutsch)

```python
%%sql 
# Your code here
SELECT lg.language_name, g.genre_name, m.title, MAX(m.popularity)
FROM languages AS lg
JOIN languagemap AS lm ON lg.iso_639_1 = lm.iso_639_1
JOIN movies AS m ON lm.movie_id = m.movie_id
JOIN genremap AS gm ON m.movie_id = gm.movie_id
JOIN genres AS g ON gm.genre_id = g.genre_id
WHERE lg.language_name LIKE "%deutsch%" AND g.genre_name = 'Action';
```

### Question 14
How many movies contain at least one of the official South African Languages, Afrikaans or Zulu?

```python
%%sql 
# Your code here
%%sql

SELECT count(DISTINCT(lm.movie_id))
FROM languagemap AS lm
INNER JOIN languages AS l
ON l.iso_639_1 = lm.iso_639_1
WHERE l.language_name LIKE "%Isizulu" or l.language_name LIKE "%Afrikaans"
;
```
